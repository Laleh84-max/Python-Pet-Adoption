```python
def printMenu(df):
    print("------------------------------------------------------------------------------\n" + \
        "-----------------------WELCOME TO THE ADOPTION APP!!!---------------------------\n" + \
         "--------------------------------------------------------------------------------\n")
    while True:
        opt = input(
        "\n" + \
        "Select an option:\n" + \
        "1) Display all animals\n" + \
        "2) Display number of animals\n" + \
        "3) Display animal by Index\n" + \
        "4) Display animal by Kind\n" + \
        "5) Calculate average age\n" + \
        "6) Create new animal\n" + \
        "7) Edit animal\n" + \
        "8) Delete an animal\n" + \
        "9) Save file\n" + \
        "0) Exit\n\n" + \
        " Option:")
        if opt.isdigit():
            opt= int(opt)
            if opt == 0: 
                exitMenu()
                break
            elif opt == 1:
                printAllAnimals(df)
            elif opt == 2:
                countAnimals(df)
            elif opt == 3:
                index_n = int(input("Introduce index:"))
                printAnimal(df, index_n)
            elif opt == 4:
                kind = input("Introduce kind: (options: cat, dog, snake, hedgehog, rabbit, " + 
                             "turtle, guinea pig, hamster, bird): ")
                printAnimalKind(df, kind)
            elif opt == 5:
                calcAvg(df)
            elif opt == 6:
                df = createAnimal(df)
                print("New record was created!!!")
            elif opt == 7:
                index_n = int(input("Introduce index of Animal to edit: "))
                column = input("What would you like to edit? (Options: Name, Kind, Age, Breed," +
                               " Vaccinated, Location, Gender, Neutered or Date Found): ")
                new_value = input("Introduce new value: ")
                editAnimal(df, index_n, column, new_value)
                print("\n\n Record was editted!!! \n\n")
            elif opt == 8:
                index_n = int(input("introduce index of a animal to delete:"))
                deleteAnimal(df, index_n)
                print("\n\n Record was deleted \n\n")
            elif opt ==9:
                saveFile(df)
                print("Changes were saved to the file!")
            else: 
                print("Please introduce a valid number!!!")
        else:
            print("please introduce a number!!")           
```

```python
def exitMenu():
    print("Thank you for using this cool app!")
```

```python
def printAllAnimals(df):
    print(df.to_string())
```

```python
def countAnimals(df):
    print("The number of animals is ", df.shape[0])
```

```python
def printAnimal(df, index_animal ):
    print("\n\n")
    print(df.loc[[index_animal]].to_string())
    print("\n\n")
```

```python
def printAnimalKind(df,kind):
    print("\n\n")
    print(df[df["kind"] == kind].to_string())
    print("\n\n")
```

```python
def calcAvg(df):
    mean = round(df["Age"].mean(),2)
    print("Average Age is", mean)
```
